```python
under_18 = data[data['Age']=='Under 18 years old']
above_18=data[data['Age']=='18-24 years old']
above_25 = data[data['Age']=='25-34 years old']
above_35 = data[data['Age']=='35-44 years old']
above_45 = data[data['Age']=='45-54 years old']
```

```python
import matplotlib.pyplot as plt
```

```python
under_18_dictionary={}
for i in under_18.index:
    keys = under_18.loc[i,'LearnCode'].split(';')
    for key in keys:
        if key in under_18_dictionary:
            val = under_18_dictionary.get(key)
            under_18_dictionary[key]=val+1
        else:
            under_18_dictionary[key]=1
```

```python
under_18_dictionary
```

```python
under_18_preferredways=list(under_18_dictionary.keys())
under_18_values = list(under_18_dictionary.values())
plt.bar(under_18_preferredways,under_18_values,ec='black')
plt.xticks(rotation=90)
plt.show()
```

For students under 18, going to school and learning from Books/Physical Media are the preferred ways to learn coding
